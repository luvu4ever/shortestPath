## Wall-follower
Thuật toán này có cách hoạt động rất đơn giản. Robot sẽ chạy theo bức tường bên trái của ma trận. Đây là các giải ma trận thô sơ mà thời nhỏ chúng ta thường sử dụng, và có thể giải được hầu hết các ma trận cơ bản. 
##### Hạn chế:
- Tuy nhiên, việc đưa điểm đích vào bên trong ma trận, và tạo các bức tường không có liên kết sẽ làm cho thuật toán này không tìm được kết quả.

## Depth First Search
Cách hoạt động: Chương trình sẽ xét các điểm được thêm vào theo trình tự mới nhất, ghi lại những điểm giao nhau (intersection). Khi đến 1 điểm và không thể tìm được lối đi tiếp (dead end), chương trình sẽ quay về điểm giao nhau và chọn một đường đi khác.
##### Hạn chế:
- Đây là một thuật toán còn khá đơn giản, và có 1 vấn đề đặt ra:
Đó là thuật toán chỉ quay lui khi đi đến một điểm dead end, do vậy không đảm bảo tìm được đường đi ngắn nhất.

## Recursive Backtracking
- Đây là thuật toán giống với thuật toán DFS, nhưng cách sử dụng lại khác nhau.
- Đối với thuật toán DFS, ta sử dụng khi đã có dữ liệu về toàn bộ ma trận, do đó có thể sử dụng chương trình để xét đường đi, quay lui và tìm đường đi ngắn nhất.
- Thuật toán Recursive Backtracking được sử dụng khi ta chưa có dữ liệu gì về ma trận, và chính robot sẽ phải thực hiện việc quay lui.
##### Hạn chế:
- Giống với thuật toán DFS, thuật toán Recursive Backtracking chưa đảm bảo tìm được đường đi tối ưu.

## Breadth First Search
Cách hoạt động: Tương tự thuật toán DFS, tuy nhiên ta sẽ xét các điểm theo trình tự thêm vào cũ nhất.
Đây là một thuật toán tối ưu, khi luôn tìm được đường đi ngắn nhất, do các điểm được xét luôn được xét theo khoảng cách ngắn nhất đến điểm đó.
##### Hạn chế:
- Tốn kém bộ nhớ: BFS yêu cầu lưu trữ các điểm ở mỗi mức sâu trong hàng đợi, điều này dẫn đến việc sử dụng bộ nhớ rất lớn. Trong trường hợp xấu nhất, bộ nhớ cần thiết là O(V) với V là số lượng điểm trong đồ thị.
- Không hiệu quả trong trường hợp khoảng cách lớn
- Đôi khi chậm hơn DFS.

## Dijktra
Cách hoạt động: Thuật toán Dijktra sẽ xét các điểm theo trọng số nhỏ nhất. Thuật toán này tối ưu hơn DFS và BFS ở điểm:
- Đối với DFS và BFS, ta không có đánh giá khi chọn điểm tiếp theo để xét. Do vậy sẽ rất dễ chọn sai điểm.
- Dijktra sẽ sử dụng trọng số của điểm xét trước cộng với khoảng cách đến điểm hiện tại. Trong trường hợp này, luôn luôn là một khi ta chỉ di chuyển được 4 hướng.
Việc sử dụng này có vẻ rất tốt trên lý thuyết, tuy nhiên sau quá trình nghiên cứu, nhóm em nhận ra rằng, cách hoạt động này không khác so với BFS. Có thể với ma trận có các chướng ngại vật, thuật toán này sẽ hoạt động tốt hơn.

## A* Search
Cách hoạt động: Thuật toán A* sẽ xét các điểm theo một hàm trọng số là tổng của: Trọng số điểm cần xét và khoảng cách từ điểm cần xét đến điểm đích theo công thức Manhattan.
Việc triển khai A* Search cơ bản tương tự như DFS hoặc BFS. Sự khác biệt chính là trong A* Search, **fringe** sử dụng hàng đợi ưu tiên và các node có mức độ ưu tiên thấp hơn sẽ được mở rộng trước vì mong muốn giảm thiểu tổng chi phí hiện tại và hàm **heuristic**.

Trong bài toán này, có hai hàm heuristic:

- Đầu tiên đánh giá khoảng cách đường thẳng từ nút hiện tại đến nút mục tiêu, được biểu diễn bằng $h_{SLD}(n)$

- Thứ hai là đánh giá khoảng cách *Manhattan* từ nút hiện tại đến nút mục tiêu, được biểu diễn bằng $h_{MD}(n)$

Heuristics thoả mãn *consistency*:

- Đối với Manhattan distance (MD):

    Hai nút liền kề bất kỳ có hàm heuristic nhỏ hơn nhau đúng 1 và chi phí giữa chúng bằng 1. Do đó, ta có bất đẳng thức tam giác: $h(n) \leq c\left(n, a, n^{\prime}\right)+h\left(n^{\prime}\right)$ đúng cho hai nút liền kề bất kỳ.

- Đối với Straight-line distance (SLD):

    Một nút tùy ý di chuyển theo chiều ngang hoặc chiều dọc nhưng không phải cả hai. Vì vậy:

    $\begin{aligned} & h(n) \leq c\left(n, a, n^{\prime}\right)+h\left(n^{\prime}\right) \\ & \Leftrightarrow \sqrt{A^2+B^2} \leq 1+\sqrt{C^2+B^2}, \text { where }|A-C|=1 \\ & \Leftrightarrow \frac{A^2-C^2}{\sqrt{A^2+B^2}+\sqrt{C^2+B^2}} \leq 1 \Leftrightarrow \frac{(A-C)(A+C)}{\sqrt{A^2+B^2}+\sqrt{C^2+B^2}} \leq 1\left(^*\right)\end{aligned}$
    
    If $A<C$ then (*) is true.
    
    Otherwise, $(A-C)(A+C)=A+C$, and $\sqrt{A^2+B^2}+\sqrt{C^2+B^2} \geq A+C$ 

    Hence, $\left({ }^*\right)$ is also true.

    Như vậy, có thể có được lời giải tối ưu khi sử dụng các hàm heuristic này.

    Tuy nhiên, đối với mọi nút, khoảng cách đường thẳng luôn nhỏ hơn hoặc bằng khoảng cách Manhattan:

    => Vì vậy, tốt hơn nên sử dụng khoảng cách **Manhattan** trong bài toán này.

Biểu đồ dưới đây biểu diễn trung bình về thời gian chạy, quãng đường tìm được và lượng ô cần tìm kiếm đối với 4 thuật toán DFS, BFS, A* và Dijktra.

Dữ liệu được lưu trong các file: results_time.csv, results_path.csv, results_search.csv.

Đây là kết quả sau 1000 lần chạy thử nghiệm trên các mazes khác nhau với size là 50x50, điểm mở đầu và kết thúc được random.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

time_df = pd.read_csv('results_time.csv')
search_df = pd.read_csv('results_search.csv')
path_df = pd.read_csv('results_path.csv')

def draw_bar_chart(df, title):
    algorithms = ["DFS", "BFS", "A*", "Dijktra"]
    column_means = df.mean()
    
    plt.figure(figsize=(10, 6))
    plt.bar(algorithms, column_means.values, color=['blue', 'green', 'red', 'purple'])
    plt.title(title)
    plt.xlabel('Algorithms')
    plt.ylabel('')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

draw_bar_chart(time_df, 'Average Run Time Comparison')
draw_bar_chart(path_df, 'Average Path Comparison')
draw_bar_chart(search_df, 'Average Searched Block Comparison')
```
